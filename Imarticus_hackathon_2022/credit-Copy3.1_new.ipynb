# Pre processing the data 

```python
#create a copy of dataframe
train_df=pd.DataFrame.copy(training_data)
```

```python
train_df.head()
```

```python
train_df.shape
```

```python
train_df.columns
```

```python
train_df.dtypes
```

```python
train_df.isnull().sum()
```

```python
for x in train_df.columns:

    if train_df[x].dtype =='object' or train_df[x].dtype =='bool':
         train_df[x].fillna(train_df[x].mode()[0],inplace=True)
```

```python
train_df.isnull().sum()
```

```python
train_df['Saving Account'] = train_df['Saving Account'].fillna(0)
```

```python
train_df.isnull().sum()
```

```python
colname=[]
for x in train_df.columns:
    if train_df[x].dtype=='object':
        colname.append(x)
colname        
```

```python
#for preprocessing the data
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
"""
-It will try storing all of the unique values from that (colname)
variable as the key and it will try associating numeric value 
along with it.
-Those number values that assign, are going to be in ascending order.
"""
for x in colname:
    train_df[x]=le.fit_transform(train_df[x])
    le_name_mapping=dict(zip(le.classes_, le.transform(le.classes_)))
    print('feature',x)
    print('mapping',le_name_mapping)
```

```python
X=train_df.drop('credit_amount', axis=1)
```

```python
X.shape
```

```python
Y=train_df['credit_amount']
```

```python
Y.shape
```

```python
Y=Y.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(X)

X=scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split

#split the data into train test

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=45)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
#create a model object
lm=LinearRegression()

#train the model
lm.fit(X_train,Y_train)

print(lm.intercept_)
print(lm.coef_)
```

```python
Y_pred=lm.predict(X_test)
Y_pred
```

```python
lm.score(X_train,Y_train) #its gives me r square val
```

```python
from sklearn.metrics import  r2_score,mean_squared_error

r2=r2_score(Y_test,Y_pred)
print("R-squared value :", r2)

rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
print("RMSE VALUE :",rmse)

adjusted_r_2=1-(1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)
print("Adjusted-R-squared value :",adjusted_r_2)
```
