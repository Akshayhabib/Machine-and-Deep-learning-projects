# preprocessing

```python
train_df.isnull().sum()
```

```python
#imputing categorical missing data with mode value

col_name=['Rain_OR_SNOW']

for value in col_name:
    train_df[value].fillna(train_df[value].mode()[0],inplace=True )
```

```python
train_df.columns
```

```python
train_df.dtypes
```

```python
#imputing numerical missing data with mean value

col_name=['Apparent Temperature (C)','Humidity', 'Wind Speed (km/h)','Wind Bearing (degrees)', 'Visibility (km)', 'Pressure (millibars)']

for value in col_name:
    train_df[value].fillna(round(train_df[value].mean(),0),inplace=True )
```

```python
train_df['Temperature (C)']=train_df['Temperature (C)'].astype(str).apply(lambda x:x.replace(',',''))
```

```python
train_df['Temperature (C)']=train_df['Temperature (C)'].apply(lambda x:x.replace('-','0'))
```

# Running the ensemble model

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import RandomForestClassifier

#creating sub models
estimators=[]


model_1=DecisionTreeClassifier(criterion="gini",random_state=10)
estimators.append(("DT",model_1))

#tuned knn
model_2=KNeighborsClassifier(n_neighbors=267,
                               metric='euclidean')
estimators.append(("knn",model_2))


model_3=RandomForestClassifier(n_estimators=100,random_state=10)
estimators.append(("RFT",model_3))

model_4=svc_model=SVC(kernel="rbf",C=10,gamma=0.1)
estimators.append(("SVC",model_4))
```

```python
#creating the ensemble moodel
ensemble=VotingClassifier(estimators)
ensemble.fit(X_train,Y_train)

Y_pred=ensemble.predict(X_test)
print(Y_pred)
```
