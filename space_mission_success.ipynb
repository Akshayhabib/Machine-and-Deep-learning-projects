# Pre processing the data

```python
#create a copy of dataframe
df=pd.DataFrame.copy(space_df)
```

```python
df.head()
```

```python
df=df.drop(["Unnamed: 0","Unnamed: 0.1","Location","Detail"],axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns = ['Company_Name', 'Datum', 'Status Rocket', 'Rocket', 'Status Mission']

df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
for value in df['Rocket']:
    print(type(value))
```

```python
df['Rocket'].unique()
```

```python
df['Rocket']=df['Rocket'].astype(str).apply(lambda x:x.replace(',','')).astype(np.float32)
```

```python
df.dtypes
```

```python
df.Rocket.mean()
```

```python
#replace the missing values with mean values
df['Rocket'].fillna(df['Rocket'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Datum']=pd.to_datetime(df['Datum'], utc=True)
df['Datum'].head()
```

```python
df["Year"]=df['Datum'].dt.year
df.head()
```

```python
df["Month"]=df['Datum'].dt.month
df.head()
```

```python
df["Day"]=df['Datum'].dt.day
df.head()
```

```python
df.drop('Datum', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Status Mission'].unique()
```

```python
df['Status Mission'] = df['Status Mission'].apply(lambda x: x if x == 'Success' else 'Failure')
```

```python
df['Status Mission'].value_counts()
```

```python
#for preprocessing the data

from sklearn.preprocessing import LabelEncoder

colname=['Status Rocket', 'Status Mission']

le=LabelEncoder()

for x in  colname:
    df[x]=le.fit_transform(df[x])
    
    le_name_mapping=dict(zip(le.classes_, le.transform(le.classes_)))
    print('feature',x)
    print('mapping',le_name_mapping)

```

```python
"""
Status Mission
 {'Failure': 0, 'Success': 1}
"""
```

```python
df.head()
```

```python
df.columns
```

```python
#Using One Hot Encoding For company name
dummies= pd.get_dummies(df.Company_Name)
dummies.head()
```

```python
df= pd.concat([df, dummies], axis=1)
df.head()
```

```python
new_df=df.drop('Company_Name',axis='columns')
new_df.head()
```

```python
Y = new_df['Status Mission']
X = new_df.drop('Status Mission', axis=1)
```

```python
X.shape
```

```python
Y.shape
```

```python
#scaling the data

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(X)

X=scaler.transform(X)

print(X)
```

```python
from sklearn.model_selection import train_test_split

#split the data  into test and train 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,
                                                 random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
#creating a model
classifier=LogisticRegression()

#fitting training data to the model
classifier.fit(X_train,Y_train)

Y_pred=classifier.predict(X_test)
#print(list(zip(Y_test,Y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)


print("CLASSIFICATION REPORT: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test,Y_pred)
print("Accuracy of the model : ",acc)
```

# adjustung the threshold

```python
#store the predicted probabilities
y_pred_prob=classifier.predict_proba(X_test)
y_pred_prob
```

```python
y_pred_class=[]

for value in y_pred_prob[:,1]:
    if value > 0.60:
        y_pred_class.append(1)
        
    else:
        y_pred_class.append(0)
        
#print(y_pred_class)  
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(Y_test,y_pred_class)
print(cfm)


print("CLASSIFICATION REPORT: ")

print(classification_report(Y_test,y_pred_class))

acc=accuracy_score(Y_test,y_pred_class)
print("Accuracy of the model : ",acc)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
X_train.shape
```

```python
inputs = tf.keras.Input(shape=(61,))
x = tf.keras.layers.Dense(20, activation='relu')(inputs)
x = tf.keras.layers.Dense(16, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[tf.keras.metrics.AUC(name='auc')]
)

batch_size=32
epochs=35

history = model.fit(
    X_train,
    Y_train,
    validation_split=0.3,
    batch_size=batch_size,
    epochs=epochs
)
```

```python
plt.figure(figsize=(14, 10))

epochs_range = range(1, epochs + 1)
train_loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(epochs_range, train_loss, label="Training Loss")
plt.plot(epochs_range, val_loss, label="Validation Loss")

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend('upper right')

plt.show()
```

```python
model.evaluate(X_test, Y_test)
```
